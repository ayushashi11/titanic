```python
def compute_cost(W, b, X, Y, regularization_strength = 10000):
    N = X.shape[0]
    distances = 1 - Y * (X@W+b)
    distances[distances<0] = 0
    hinge_loss = regularization_strength * (np.sum(distances) / N)
    cost = 1 / 2 * (W@W.T) + hinge_loss
    return cost

def calculate_cost_gradient(W, b, X_batch, Y_batch, regularization_strength = 10000):
    if type(Y_batch) == np.float64:
        Y_batch = np.array([Y_batch])
        X_batch = np.array([X_batch])

    distance = 1 - (Y_batch * (X_batch@W + b))
    distance[distance<0] = 0
    # dw = W - (regularization_strength * Y_batch * X_batch) * np.sign(np.abs(distance))
    dw = W - (regularization_strength * Y_batch[0] * X_batch) * np.sign(distance[0])
    return dw

def calculate_bias_cost_gradient(W, b, X_batch, Y_batch, regularization_strength = 10000):
    if type(Y_batch) == np.float64:
        Y_batch = np.array([Y_batch])
        X_batch = np.array([X_batch])

    distance = 1 - (Y_batch * (X_batch@W + b))
    distance[distance<0] = 0
    db = 0
    # dw = W - (regularization_strength * Y_batch * X_batch) * np.sign(np.abs(distance))
    db += - (regularization_strength * Y_batch) * np.sign(distance[0])
    return db

def sgd(weights, bias, X, Y, max_epochs = 5000, cost_threshold = 0.00001, alpha = 0.0000001):
    if weights.shape==(0,):
        weights = np.zeros(X.shape[1])
    nth = 0
    prev_cost = float("inf")
    for epoch in range(1, max_epochs):
        Xt, Yt = shuffle(X, Y)
        for ind, x in enumerate(Xt):
            error = calculate_cost_gradient(weights, bias, x, Yt[ind])
            bias_error = calculate_bias_cost_gradient(weights, bias, x, Yt[ind])
            weights = weights - (alpha * error)
            bias = bias - alpha * bias_error
        if epoch == 2 ** nth or epoch == max_epochs - 1:
            cost = compute_cost(weights, bias, X, Y)
            print("Epoch:",epoch,"Cost:",cost)
            if abs(prev_cost - cost) < cost_threshold * prev_cost:
                return (weights, bias, cost)
            prev_cost = cost
            nth += 1
    return (weights, bias, prev_cost)

def predict(weights, bias, X):
    p=(X@weights+bias)
    p[p>0]=1
    p[p==0]=random.choice([1,0])
    p[p<0]=0
    return p
```

```python
lrweights = np.array([[]])
lrbias = np.array([[]])
lrY = np.array(list(map((lambda x: [1,0] if x==1 else [0,1]), Y)))
#activation
def sigmoid(W,b,X):
    Z=X@W.T+b
    return 1/(1+np.exp(-Z))
# cost function
def total_cost(weights, biases, x, y):
    p=sigmoid(weights, biases, x)
    return -np.sum(y*np.log(p)+(1-y)*np.log(1-p)) / len(y[0])

# gradient descent
def fit(weights, biases, x, y, alpha=0.00000001, iters=2000, cost_threshold = 0.00001):
    DATALEN: int = len(y)
    NOUT: int = len(y[0])
    NVARS: int = len(x[0])
    if weights.shape==(1,0):
        weights = np.zeros((NOUT, NVARS))
    if biases.shape==(1,0):
        biases = np.zeros((1, NOUT))
    history = []
    n=0
    prev_cost = float("inf")
    for i in range(iters):
        xt, yt = shuffle(x, y)
        p = sigmoid(weights, biases, xt)
        error = p - yt
        weights -= (error.T@xt / DATALEN) * alpha
        biases -= (np.sum(error) / DATALEN) * alpha
        if i == 2 ** n:
            cost = total_cost(weights, bias, X, Y)
            print("Epoch:",i,"Cost:",cost)
            history.append([i,cost])
            if abs(prev_cost - cost) < cost_threshold * prev_cost:
                return weights, bias, history
            prev_cost = cost
            n += 1
        history.append([i, total_cost(weights, biases, xt, yt)])
    return weights, biases, history

# prediction
def lrpredict(weights, biases, x):
    return sigmoid(weights, biases, x)

def r_squared(prediction, values):
    sstot = np.sum((values-np.mean(values))**2)
    ssres = np.sum((values-prediction)**2)
    return 1-ssres/sstot
```
